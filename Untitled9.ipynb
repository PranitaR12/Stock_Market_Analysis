```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
```

```python
# Load stock data for Apple (AAPL) from Yahoo Finance
stock_data = yf.download('AAPL', start='2020-01-01', end='2022-01-01')
stock_data['Close'].plot(title='AAPL Closing Prices')
plt.show()
```

```python
# Calculate moving averages
stock_data['MA_10'] = stock_data['Close'].rolling(window=10).mean()
stock_data['MA_50'] = stock_data['Close'].rolling(window=50).mean()

# Plot the moving averages
plt.figure(figsize=(10,6))
plt.plot(stock_data['Close'], label='AAPL Close')
plt.plot(stock_data['MA_10'], label='10-Day MA')
plt.plot(stock_data['MA_50'], label='50-Day MA')
plt.legend()
plt.title('Moving Averages')
plt.show()
```

```python
# Prepare data for ARIMA model
stock_data['Returns'] = stock_data['Close'].pct_change()
stock_data.dropna(inplace=True)

# Fit ARIMA model
model = ARIMA(stock_data['Close'], order=(5, 1, 0))  # Example order (p,d,q)
arima_model = model.fit()

# Make predictions
stock_data['ARIMA_Predictions'] = arima_model.predict(start=len(stock_data)-100, end=len(stock_data)-1, typ='levels')

# Plot predictions
plt.figure(figsize=(10,6))
plt.plot(stock_data['Close'], label='AAPL Close')
plt.plot(stock_data['ARIMA_Predictions'], label='ARIMA Predictions', color='red')
plt.legend()
plt.title('ARIMA Model Predictions')
plt.show()
```
